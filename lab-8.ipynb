```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/ADANIPORTS.csv')

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

start_date = '2007-11-27'
end_date = '2008-02-06'

filtered_data = df.loc[start_date:end_date]

plt.figure(figsize=(10, 6))
plt.plot(filtered_data.index, filtered_data['Open'], label='Opening Price', color='blue')
plt.plot(filtered_data.index, filtered_data['Close'], label='Closing Price', color='aqua')

plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Stock Prices between {} and {}'.format(start_date, end_date))

plt.legend()

plt.show()
```
