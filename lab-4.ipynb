**4 a) To add, subtract, multiply, divide arguments element-wise**

```python
import numpy as np

print("Addition --> ",end ="")
print(np.add(3.0,3.0))

print("Subtraction --> ",end="")
print(np.subtract(18.0,9.0))

print("Multiplication --> ",end="")
print(np.multiply(3.0,5.0))

print("Division --> ",end="")
print(np.divide(10.0,5.0))
```

**4 b) To round elements of the array to the nearest integer**

```python
import numpy as np

arrayy = np.array([0.2,0.7,1.3,1.5])
print("Original array --> ",end="")
print(arrayy)

arrayy = np.rint(arrayy)
print("After rounding array --> ",end="")
print(arrayy)
```

**4 c) To calculate mean across dimension, in a 2D numpy array**

```python
import numpy as np

array = np.array([[40,50],[40,50]])
print("Original array -->\n",array)

arrayX = np.mean(array,axis=0)
print("Mean of X-axis --> ",arrayX)

arrayY = np.mean(array,axis=1)
print("Mean of Y-axis --> ",arrayY)
```

**4 d) To calculate the difference between neighboring elements, element-wise of a given array**

```python
import numpy as np

arrayy = np.array([5,9,29,38,77,65,143,130])
print("Original array --> ",arrayy)

result = np.diff(arrayy)
print("Result --> ",result)
```
