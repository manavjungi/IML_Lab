**5 a) To find the maximum and minimum value of a given flattened array**

```python
import numpy as np

arrayy = np.array([[20,55,84],[54,23,20]])

print("Original array --> \n",arrayy)

print("Maximum value --> ",np.amax(arrayy))

print("Minimum value --> ",np.amin(arrayy))
```

**5 b) To compute the mean, standard deviation, and variance of a given array along the second axis.**

```python
import numpy as np

arrayy = np.array([[33,22,11],[44,55,66]])

print("Original array -->\n",arrayy)

meanValue = np.mean(arrayy, axis = 1)
print("Mean value --> ",meanValue)

stdValue = np.std(arrayy, axis = 1)
print("Standard deviation --> ",stdValue)

variValue = np.var(arrayy, axis = 1)
print("Variance --> ",variValue)
```
